# 简单绘图
## 绘制多项式函数

```python
import numpy as np
import matplotlib.pyplot as plt

func = np.poly1d(np.array([1, 2, 3, 4]).astype(float))
x = np.linspace(-10, 10 ,30)
y = func(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y(x)')
plt.show()
```

## 绘制多项式函数及其导函数

```python
import numpy as np
import matplotlib.pyplot as plt

func = np.poly1d(np.array([1, 2, 3, 4]).astype(np.float64))
func1 = func.deriv(m=1)
x = np.linspace(-10, 10, 30)
y = func(x)
y1 = func1(x)

plt.plot(x, y, 'ro', x, y1, 'g--')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# 子图

使用 subplot 函数创建第一个子图。该函数的第一个参数是子图的行数，第二个参数是子图的列数，第三个参数一个从1开始的序号。另一种方式是将这3个参数结合成一个数字，如311。

```python
import numpy as np
import matplotlib.pyplot as plt

func = np.poly1d(np.array([1, 2, 3, 4]).astype(np.float64))
x = np.linspace(-10, 10, 30)
y = func(x)
func1 = func.deriv(m=1)
y1 = func1(x)
func2 = func.deriv(m=2)
y2 = func2(x)

plt.subplot(311)
plt.plot(x, y, 'r--')
plt.title('Polynomial')
plt.subplot(312)
plt.plot(x, y1, 'b^')
plt.title('First Derivative')
plt.subplot(313)
plt.plot(x, y2, 'go')
plt.title('Second Derivateive')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
